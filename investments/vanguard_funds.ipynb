```python
start_date = '1990-01-01'
end_date = '2023-01-01'
```

```python
prices_array = []
for symbol in vanguard_funds["SYMBOL"]:
    data = yf.download(symbol, start_date, end_date)
    if not len(data):
        continue
    print(f"Loading ... {symbol}")
    prices_array.append(data[["Adj Close"]].rename(columns={"Adj Close": symbol}))
prices = pd.concat(prices_array, axis=1)
```
